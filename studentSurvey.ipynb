```python
df = pd.read_csv("a2.csv")
df.columns = ["followers", "phone_time", "energy", "music_training", "mom_height", "student_height"]
df.head()
```

```python
# get rid of commas and turn the followers column into integers
df["followers"] = df["followers"].apply(lambda x : int(x.replace(",", "")))
```

```python
df.followers.dtype
```

```python
scatter = sns.scatterplot(data=df, x="followers", y="phone_time")
plt.title("Relationship between Instagram followers & screen time")
plt.xlabel("Instagram followers")
plt.ylabel("Screen time (min)")
plt.show()
```

```python
bar = sns.barplot(data=df, x="energy", y="followers", ci=None, palette="Blues")
plt.title("Average Instagram Followers for Introverts vs. Extroverts")
plt.xlabel("")
plt.ylabel("Instagram followers")
plt.show()
```

```python
bar2 = sns.barplot(data=df, x="energy", y="music_training", ci=None, palette="Blues")
plt.title("Average Music Training for Introverts vs. Extroverts")
plt.xlabel("")
plt.ylabel("Music Training (years)")
plt.show()
```
